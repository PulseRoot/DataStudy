```python
df_raw = pd.read_csv("국가_대륙_별_상품군별_온라인쇼핑_해외직접판매액_20210809140812.csv", encoding="cp949")
```

#분석과 시각화를 위한 tidy data 만들기

```python
df = df_raw.melt(id_vars=["국가(대륙)별", "상품군별", "판매유형별"], var_name="기간", value_name="백만원") #리스트 형태로 선언해야함 
#컬럼 중에 몇개 만 id_vars로 지정하면 나머지 모든 컬럼은 value_vars의 소속이 된다.
df.shape
```

```python
df.head()
```

```python
df.info()
```

#데이터 전처리

#기간에서 연도 분리

```python
int ("2019 4/4 p)".split()[0])
```

```python
df["연도"] = df["기간"].map(lambda x : int (x.split()[0])) #숫자데이터만 분리 + int로 자료형 변경
df.head()
```

```python
#분기만 분리하기
"2019 1/4 p)".split()[1].split("/")[0]
```

```python
df["분기"] = df["기간"].map(lambda x:int(x.split()[1].split("/")[0]))
df.head()  
```

#금액을 수치데이터로 변경하기 위한 자료형 변환 

```python
df["백만원"] = df["백만원"].replace("-", pd.np.nan).astype(float)
df.head()  #결측치로 변경 데이터 타입 변경을 위해 필요함
```

#필요없는 데이터 제거하기

```python
df.info()
```

```python
df = df[(df["국가(대륙)별"] != "합계") & (df["상품군별"] != "합계")].copy() #필요없는 데이터 삭제
```

```python
df.info()
```

```python
#결측치 보기
df.isnull().sum()
```

# K-beauty 시각화

```python
df_total = df[df["판매유형별"] == "계"].copy()
df_total.head()
```

```python
sns.lineplot(data=df_total, x="연도", y="백만원")
```

```python
sns.lineplot(data=df_total, x="연도", y="백만원", hue="상품군별") #상품군별로 다른 색으로
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.) #레전드 값 = 상품군별 색상을 나타내는 박스를 그래프 밖으로 옮기는 코드
```

```python
#서브플롯으로 그리기
sns.relplot(data=df_total,x="연도", y="백만원", col="상품군별", col_wrap=4
            ,hue="상품군별", kind="line")# col="상품군별", col_wrap=4을 통해 상품군별로 그래프를 그려줌
```

```python
df_sub = df_total[~df_total["상품군별"].isin(["화장품", "의류 및 패션관련 상품"])].copy()  #어느 한 카테고리만 보기 ~를 붙히면 그 값만 빠진 값들이 보임
```

```python
sns.relplot(data=df_sub,x="연도", y="백만원", col="상품군별", col_wrap=4
            ,hue="상품군별", kind="line")
```
